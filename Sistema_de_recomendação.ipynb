Criando funções necessarias para a recomendação

```
def carregarBases(): 
    filmes = { }
    arquivo = open('movies.csv')
    data = [r for r in arquivo]
    header = data[0]
    data = data[1:]

    for linha in data:
        (id,titulo) = linha.split(',') [0:2] 
        filmes[id] = titulo
    
    base = { }
    arquivo2 = open('ratings.csv')
    data2 = [r for r in arquivo2]
    header2 = data2[0]
    data2 = data2[1:]
    for linha in data2:
        (usuario,idfilme,nota,tempo) = linha.split(',')
        base.setdefault(usuario,{ })
        base[usuario][filmes[idfilme]] = float(nota)

    return base
```

```
from math import sqrt

def getAproximacao(base,user1,user2):  # Calculo Eclidiano
    dados = {}

    for i in base[user1]:
        if i in base[user2]:
            dados[i] = 1

    if len(dados) == 0:
        return 0

    soma = sum ([pow(base[user1][i] - base[user2][i],2)
                    for i in base[user1] if i in base[user2]
                ])
    return 1/(1+sqrt(soma))
```

```
def getRecomendacoes(base,user):
    totais = {}
    somasimilaridade = {}

    for newUser in base: 
        if newUser == user:
            continue

        similaridade = getAproximacao(base,user,newUser)

        if similaridade <= 0:
            continue

        for item in base[newUser]: 
            if item not in base[user]: 
                totais.setdefault(item,0)
                totais[item] += base[newUser][item] * similaridade
                somasimilaridade.setdefault(item,0)
                somasimilaridade[item] += similaridade

    rankings = [(total/somasimilaridade[item],item) for item , total in totais.items()]
    rankings.sort()
    rankings.reverse()
    return rankings
```

```
def getSimilares(base,user): 
    similaridade = [(getAproximacao(base,user, outro), outro)
                    for outro in base if outro != user]
    similaridade.sort()
    similaridade.reverse()
    return similaridade
```
